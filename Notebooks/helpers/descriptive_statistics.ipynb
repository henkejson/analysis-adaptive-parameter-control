```python
import pandas as pd

# Load the CSV file
data = pd.read_csv("../../single_parameter/combined_data/statistics.csv")

# Group the data by 'TargetModule'
grouped = data.groupby('TargetModule')

def calculate_stats(group, column):
    return {
        'Mean': group[column].mean(),
        'Min': group[column].min(),
        'Max': group[column].max(),
        '25%': group[column].quantile(0.25),
        '75%': group[column].quantile(0.75),
        'Std Dev': group[column].std()
    }

# Apply the function to each group for each metric, with include_groups=False
coverage_stats = grouped.apply(calculate_stats, column='Coverage', include_groups=False).apply(pd.Series)
algorithm_iterations_stats = grouped.apply(calculate_stats, column='AlgorithmIterations', include_groups=False).apply(pd.Series)

# Print the results
print("Coverage Statistics by TargetModule:")
print(coverage_stats)
print("\nAlgorithm Iterations Statistics by TargetModule:")
print(algorithm_iterations_stats)
# Optionally, save the results to a new CSV if needed
# stats_result.to_csv('output_statistics.csv')
```
